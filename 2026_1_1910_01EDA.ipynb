```python
#https://pastebin.com/nNgDpmMr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.data
```

```python
print (iris.DESCR)
```

```python
iris_df = pd.DataFrame(data= iris.data, columns= iris.feature_names)

iris_df['species_num'] = iris.target

iris_df['species']= pd.Categorical.from_codes(iris.target,iris.target_names)

iris_df.head()
```

```python
num_cols=['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

# Variables categoricas
cat_cols = 'species'
```

```python
#analisis de variable
for i in num_cols:
    sns.displot(data=iris_df, x=i, hue='species', kde=True)
    plt.show()
```

```python
#analisis bivariado

for i in num_cols:
    (pd.crosstab(iris_df[i], iris_df['species'], normalize='index')*100).plot(kind='bar', stacked=True)
    plt.show()
```

```python

#Boxplot Bivariado

for i in num_cols:

   sns.boxplot(x='species', y=i, data=iris_df, palette='tab10', hue='species')

   plt.title(f'Boxplot {i}')
   plt.ylabel(i)

   plt.xlabel('Species')

   plt.show()
```

```python
plt.figure(figsize=(12, 8))

j=1

for i in num_cols:

    plt.subplot(2, 2, j)

    sns.boxplot(x='species', y=i, data=iris_df, palette='tab10', hue='species')

    plt.title(f'Boxplot {i}')

    plt.ylabel(i)

    plt.xlabel('Species')

    j+=1

plt.show
```

```python
plt.figure(figsize=(12, 8))

j=1

for i in num_cols:

    plt.subplot(2, 2, j)

    sns.boxplot( y=i, data=iris_df)

    plt.title(f'Boxplot {i}')

    plt.ylabel(i)

    j+=1

plt.show()
```

```python
#Distribucion de variables categoricas

print(iris_df['species'].value_counts())

sns.countplot(data=iris_df, x='species')

plt.show()
```

```python
#Media de las variables numÃ©ricas y la clase
iris_df.groupby('species', observed= False).mean().T
```

```python

#https://pastebin.com/nNgDpmMr

#Grafico  de distribucion Sepal - Longitud vs Ancho
# Definir el tamaÃ±o de la figura
plt.figure(figsize=(10, 6))

# Graficar cada especie
for species, color in zip(iris_df['species'].unique(), ['blue', 'green', 'red']):
    subset = iris_df[iris_df.species == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species.capitalize(), color=color, alpha=0.7)

# AÃ±adir etiquetas y tÃ­tulo
plt.xlabel('SÃ©palo - Longitud (cm)')
plt.ylabel('SÃ©palo - Ancho (cm)')
plt.title('AnÃ¡lisis de SÃ©palo - Longitud vs Ancho por Especie')

# AÃ±adir una leyenda
plt.legend(title='Especie')

# AÃ±adir una cuadrÃ­cula para mejor lectura
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el grÃ¡fico
plt.show()
```
